## Aim

In this notebook, I evenly and randomly splited the affiliation task. 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/processed/research_author_with_pred.csv')
df['aff_prediction'] = np.nan
df['country_code'] = np.nan
df['aff_type'] = np.nan 
```

```python
# these 138 authors are scattered in 100 papers
len(list(set(df[df['affiliation.2'].isnull()].doi)))
```

```python
df = df[(df.authorFullName.notnull()) & (df['affiliation.2'].notnull())]
df.shape
```

```python
# d = dict(zip(df['affiliation.2'], df.ROR_AFFNAME))
# len(d.items())
# pd.DataFrame(d.items())
```

```python
shuffled = df.sample(frac = 1, random_state = 200)
result = np.array_split(shuffled, 2)
```
